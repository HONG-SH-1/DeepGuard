```python
import numpy as np
import tensorflow as tf

PREP = '/kaggle/input/notebooks/hongsh184/test122/preprocessed/'
MODELS = '/kaggle/input/notebooks/hongsh184/test1222/models/'

X_te_tue_w5 = np.load(PREP + 'X_te_tue_w5.npy')
X_te_wed_w5 = np.load(PREP + 'X_te_wed_w5.npy')
X_te_fri_w5 = np.load(PREP + 'X_te_fri_w5.npy')
y_te_tue_w5 = np.load(PREP + 'y_te_tue_w5.npy', allow_pickle=True)
y_te_wed_w5 = np.load(PREP + 'y_te_wed_w5.npy', allow_pickle=True)
y_te_fri_w5 = np.load(PREP + 'y_te_fri_w5.npy', allow_pickle=True)

def evaluate(model_path, threshold_path, name):
    model = tf.keras.models.load_model(model_path)
    threshold = float(np.load(threshold_path))
    results = {}
    for X, y, label in [
        (X_te_tue_w5, y_te_tue_w5, 'BruteForce'),
        (X_te_wed_w5, y_te_wed_w5, 'DoS'),
        (X_te_fri_w5, y_te_fri_w5, 'DDoS'),
    ]:
        recon = model.predict(X, verbose=0, batch_size=256)
        mse = np.mean(np.power(X - recon, 2), axis=(1,2))
        pred = (mse > threshold).astype(int)
        true = (y != 'BENIGN').astype(int)
        tp = np.sum((pred==1) & (true==1))
        fn = np.sum((pred==0) & (true==1))
        recall = tp / (tp + fn + 1e-9)
        results[label] = recall
    avg = np.mean(list(results.values()))
    print(f"\n{name} (threshold={threshold:.6f})")
    print(f"  BruteForce: {results['BruteForce']:.3f}")
    print(f"  DoS:        {results['DoS']:.3f}")
    print(f"  DDoS:       {results['DDoS']:.3f}")
    print(f"  평균Recall: {avg:.3f}")

models = [
    ('mlp_model.keras',             'mlp_threshold.npy',                  'BASE (u64, p95)'),
    ('TUN1_u128_p95_nodrop.keras',  'TUN1_u128_p95_nodrop_threshold.npy', 'TUN1 (u128, p95)'),
    ('TUN2_u64_p90_nodrop.keras',   'TUN2_u64_p90_nodrop_threshold.npy',  'TUN2 (u64, p90) ✅'),
    ('TUN3_u128_p90_nodrop.keras',  'TUN3_u128_p90_nodrop_threshold.npy', 'TUN3 (u128, p90)'),
    ('TUN4_u64_p95_drop01.keras',   'TUN4_u64_p95_drop01_threshold.npy',  'TUN4 (u64, p95, drop)'),
]

for model_file, threshold_file, name in models:
    evaluate(MODELS + model_file, MODELS + threshold_file, name)
```
